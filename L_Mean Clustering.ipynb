```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data={'Math':[35, 40, 90, 88, 60, 62, 25, 27],
      'Science':[30, 25, 95, 92, 65, 70, 30, 28]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
x=df
```

```python
kmeans=KMeans(n_clusters=2,random_state=0)
kmeans.fit(df)
```

```python
df['Cluster']=kmeans.labels_
```

```python
df
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df['Math'],df['Science'],c=df['Cluster'],cmap='rainbow', s=100)
plt.xlabel('Math Marks')
plt.ylabel('Science Marks')
plt.title('K-Means Clustering: Student Performance')
plt.show()
```
