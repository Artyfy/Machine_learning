```python
import pandas as pd
import numpy as np
```

```python
#positive Skewness
Marks=[50,52,53,54,55,90,95]
Marks
```

```python
df=pd.DataFrame(Marks)
```

```python
df.skew()
```

```python
#symmetric Skewness
list=[2,3,5,77,90,39,67]
df=pd.DataFrame(list)
```

```python
df.skew()
```

```python
# negative Skewness
df = pd.DataFrame({
    "Salary": [100, 105, 110, 115, 50, 40, 30]
})

print(df.skew())
```

```python
import matplotlib.pyplot as plt
```

```python
marks=[40,45,50,55,60,65,70]
df=pd.DataFrame(marks, columns=['marks'])
print(df.skew())


plt.hist(df['marks'],bins=5,color="red",edgecolor='black')
plt.title("Skewness Example")
plt.show()
```

```python
salary = [20000, 22000, 25000, 28000, 30000, 100000, 200000]
df2 = pd.DataFrame(salary, columns=["Salary"])
print("Salary Skewness:", df2.skew())

plt.hist(df2["Salary"], bins=5, color="lightgreen", edgecolor="black")
plt.title(f"Positive Skew Example")
plt.show()
```
